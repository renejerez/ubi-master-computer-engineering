# Scraping Scores and Number of Reviews

This script automates the process of scraping review scores and star ratings from a list of URLs stored in a CSV file. Using Selenium WebDriver, it navigates to each URL, handles potential pop-ups, and extracts relevant content using BeautifulSoup. The extracted data includes review scores and star ratings, which are then stored in a DataFrame for further analysis.

## Setup and Load Data

### Import libraries

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
import re
```

### Load bodyshops from CSV file

```python
# Load the CSV file
df = pd.read_csv('bodyshops_to_check.csv')
```

### Find URLs for each bodyshop

```python
# Step 1: Set up Selenium WebDriver
def setup_driver():
    options = Options()
    options.add_argument('--disable-gpu')  # Disable GPU acceleration
    options.add_argument('--no-sandbox')  # Bypass OS security model
    options.add_argument('--disable-dev-shm-usage')  # Overcome limited resource problems
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    return driver

# Step 2: Open URL and handle pop-ups
def open_url_and_handle_popups(driver, url):
    driver.get(url)
    time.sleep(5)  # Wait for the page to load

    try:
        # Accept cookies pop-up
        accept_cookies_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='Accept all']"))
        )
        accept_cookies_button.click()
        print("Accepted cookies")
    except Exception as e:
        print(f"No cookies button found: {e}")

    return driver

# Step 3: Perform search and get URL
def search_and_get_url(driver, address, name):
    search_box = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "searchboxinput"))
    )
    search_box.clear()
    search_box.send_keys(address)
    search_box.send_keys(Keys.ENTER)
    time.sleep(5)  # Wait for search results to load

    search_box.clear()
    search_box.send_keys(name)
    search_box.send_keys(Keys.ENTER)
    time.sleep(5)  # Wait for search results to load

    current_url = driver.current_url
    return current_url

# Initialize the WebDriver
driver = setup_driver()
driver = open_url_and_handle_popups(driver, "https://www.google.com/maps")

# Iterate through the CSV and perform searches
urls = []
for index, row in df.iterrows():
    address = row['bodyshop_address']
    name = row['nome_prestador_oficina']
    url = search_and_get_url(driver, address, name)
    urls.append(url)

# Add the URLs to the DataFrame
df['url'] = urls

# Save the updated DataFrame back to the same CSV file
df.to_csv(csv_file_path, index=False)

# Close the WebDriver
driver.quit()
```

## WebDriver Configuration

### Configure and initialize Chrome WebDriver

```python
# Step 1: Set up Selenium WebDriver
def setup_driver():
    options = Options()
    options.add_argument('--disable-gpu')  # Disable GPU acceleration
    options.add_argument('--no-sandbox')  # Bypass OS security model
    options.add_argument('--disable-dev-shm-usage')  # Overcome limited resource problems
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    return driver

# Step 2: Open URL and handle pop-ups
def open_url_and_handle_popups(driver, url):
    driver.get(url)
    time.sleep(5)  # Wait for the page to load

    try:
        # Accept cookies pop-up
        accept_cookies_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='Accept all']"))
        )
        accept_cookies_button.click()
        print("Accepted cookies")
    except Exception as e:
        print(f"No cookies button found: {e}")

    return driver

# Step 3: Extract page content and find the score
def extract_score(driver):
    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Locate the score element using aria-label containing "stars"
    score_element = soup.find('span', {'aria-label': lambda x: x and 'stars' in x})
    score = "Score element not found"
    if score_element:
        score = score_element['aria-label']
        print(f"Score (Rating): {score}")
    return score

# Step 4: Extract elements containing 'stars'
def extract_star_elements(driver):
    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Locate all elements containing 'stars' in aria-label
    star_elements_info = []
    star_elements = soup.find_all(attrs={'aria-label': lambda x: x and 'stars' in x})
    for element in star_elements:
        star_elements_info.append(f"Element tag: {element.name}, Classes: {element.get('class')}, Aria-label: {element['aria-label']}")
    return star_elements_info
```

## Web Scraping and Data Extraction

### Open URLs, handle pop-ups, extract scores and 'stars' elements


```python

# Main execution
if __name__ == "__main__":
    urls = df['url'].tolist()
    results = []

    driver = setup_driver()
    for url in urls:
        if pd.isna(url) or url.strip() == "":
            results.append("No URL provided")
            continue

        driver = open_url_and_handle_popups(driver, url)
        
        # Extract the score
        score = extract_score(driver)
        
        # Extract star elements information
        star_elements_info = extract_star_elements(driver)
        
        # Combine score and star elements into a single result
        result = f"Score: {score}\n" + "\n".join(star_elements_info)
        
        # Append the data
        results.append(result)
        
        # Keep the browser open for a while to see the result
        time.sleep(10)
    
    # Close the browser
    driver.quit()

    # Add the results to the DataFrame
    df['result'] = results
    
    print("Done")
```

## Data Processing and Storage

### Parse and count star ratings

```python
# Function to extract star ratings
def extract_star_ratings(result):
    star_counts = {"1 stars": 0, "2 stars": 0, "3 stars": 0, "4 stars": 0, "5 stars": 0}
    matches = re.findall(r'(\d(?:\.\d)?) stars,? (\d+) reviews?', result)
    for match in matches:
        stars = match[0]
        count = int(match[1].replace(',', ''))
        if stars.startswith('1'):
            star_counts["1 stars"] += count
        elif stars.startswith('2'):
            star_counts["2 stars"] += count
        elif stars.startswith('3'):
            star_counts["3 stars"] += count
        elif stars.startswith('4'):
            star_counts["4 stars"] += count
        elif stars.startswith('5'):
            star_counts["5 stars"] += count
    return star_counts
```

### Store results in DataFrame

```python
# Apply the function to the dataframe
df_star_ratings = df['result'].apply(extract_star_ratings).apply(pd.Series)

# Combine the original dataframe with the new star ratings
df = pd.concat([df, df_star_ratings], axis=1)
```

## Final Data Integration

### Merge extracted scores and reviews

```python
# Function to extract score and review counts
def extract_scores_and_reviews(result):
    matches = re.findall(r'(\d+\.\d) stars (\d+) Reviews?', result)
    return [{"score": float(match[0]), "review": int(match[1].replace(',', ''))} for match in matches]

# Apply the function to extract score and review counts
extracted_data = df['result'].apply(extract_scores_and_reviews)

# Flatten the extracted data and create a new DataFrame
flattened_data = [{"url": row["url"], "score": item["score"], "review": item["review"]} for idx, row in df.iterrows() for item in extracted_data[idx]]
df_scores_reviews = pd.DataFrame(flattened_data)

# Merge the scores and reviews back to the original dataframe on the 'url' column
df = pd.merge(df, df_scores_reviews, on="url", how="left")


# Save the final dataframe to a new CSV file
df.to_csv('bodyshops_to_check_reviews.csv', index=False)

print("Data extraction complete")
```
