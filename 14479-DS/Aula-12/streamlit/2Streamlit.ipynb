## Finances

Neste exemplo vamos usar a biblioteca `yfinance` (Yahoo Finance) para junto com o `streamlit` desenhar uma demo acerca de dados financeiros da Google. [source](https://towardsdatascience.com/how-to-get-stock-data-using-python-c0de1df17e75). Comece por instalar a biblioteca financeira:

````python
!pip install yfinance
````

Para entender o funcionamento da biblioteca execute o seguinte código. A função `yf.Ticker('GOOGL')` retorna os dados financeiros da Google e guarda-os na variável `tickerData`. Posteriormente a biblioteca recorre ao método interno `history` para criar um dataframe `ticker_DF`.

````python
import yfinance as yf
from datetime import datetime

def finance(ticker):
    #get data on this ticker
    tickerData = yf.Ticker(ticker)

    #get the historical prices for this ticker
    now = datetime.today().strftime('%Y-%m-%d')

    ticker_DF = tickerData.history(period='1d', start='2010-5-31', end=now)
    
    return ticker_DF

ticker_DF = finance('GOOGL')
ticker_DF
````

Se apenas pretender ter acesso aos dados de uma única coluna proceda da seguinte forma:

````python
ticker_DF["Close"]
````

Depois de entender o funcionamento da biblioteca financeira pode agora entender o código do streamlit, cujo import é feito a partir do código `import streamlit as st`. De seguida, a função `write` permite imprimir texto na demo. A função `line_chart` cria dois plots de linhas a partir dos valores de duas colunas do dataframe. Copie o seguinte código para um ficheiro `finance.py` e execute-o a partir da prompt com o seguinte código `streamlit run finance.py`.

```
import streamlit as st
import yfinance as yf
from datetime import datetime

def finance(ticker):
    #get data on this ticker
    tickerData = yf.Ticker(ticker)

    #get the historical prices for this ticker
    now = datetime.today().strftime('%Y-%m-%d')

    ticker_DF = tickerData.history(period='1d', start='2010-5-31', end=now)
    
    return ticker_DF

ticker_DF = finance('GOOGL')

st.write("""
# Simple Stock Price App
Shown are the stock closing price and volume of Google!
""")

st.line_chart(ticker_DF["Close"])
st.line_chart(ticker_DF["Volume"])
```

## Crime

No próximo código ([credits](https://towardsdatascience.com/how-to-deploy-your-data-science-as-web-apps-easily-with-python-955dd462a9b5)) vamos explorar os dados de um ficheiro csv relativo à ocorrência de crimes em Washington e a partir deles desenvolver uma demo. Para começar recorremos `read_csv` para obter e colocar em formato pandas dataframe um ficheiro csv (com os dados necessários) que se encontra disponível a partir do seguinte link: https://query.data.world/s/6joi7hjgjmwifhl2clpldwm36xmvmx

Na posse do dataframe, aplicamos o código `to_datetime` para converter em formato datetime os valores da coluna `REPORTDATETIME`. Usamos também o parâmetro opcional `infer_datetime_format=True` que permite detetar/inferir o formato da data. Adicionalmente recorremos ao atributo `dt.day`, `dt.month` e `dt.hour` para obtermos informação específica das datas.

````python
import pandas as pd

pd.set_option('display.max_columns', None)

def load_data():
    df = pd.read_csv("https://query.data.world/s/6joi7hjgjmwifhl2clpldwm36xmvmx")
    df["REPORTDATETIME"] = pd.to_datetime(df["REPORTDATETIME"], infer_datetime_format=True)
    df["Day"] = df["REPORTDATETIME"].dt.day 
    df["Month"] = df["REPORTDATETIME"].dt.month
    df["Hour"] = df["REPORTDATETIME"].dt.hour
    return df

crime_df = load_data()
crime_df
````

Podemos também fazer uso da biblioteca `plotly` para plotar num mapa os valores do dataframe anterior. Em especifico, alimentamos a componente `lat` (latitude) com os valores `Y` do dataframe e a `lon` com os valores `X` do dataframe.

````python
import plotly.express as px

def display_map(df):
    fig = px.scatter_mapbox(df, lat="Y", lon="X", color="METHOD", zoom=10, height=300)
    fig.update_layout(mapbox_style="open-street-map")
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    return fig

display_map(crime_df)
````
