### Requests

Para fazermos download do conteúdo de um ficheiro html (ou mesmo de um ficheiro txt ou de uma imagem que esteja guardada num servidor web) recorremos ao módulo `requests` e ao método `GET`. `Requests` é um dos packages de Python mais procurados.

### BeautifulSoup

Com o código acima tivemos oportunidade de proceder ao download do código HTML de uma página web.

No desenvolvimento web, o termo "tag soup" refere-se a código HTML sintáticamente incorreto ou mal estruturado. O que o  Beautiful Soup faz de melhor é tornar o "tag soup" "beautiful" e extrair informação com facilidade.

#### Initialize BeautifulSoup Object

Para extrairmos (parse) os conteúdos pretendidos vamos recorrer ao módulo BeautifulSoup. No exemplo abaixo vamos usar este módulo para obter texto a partir da tag `p` de uma dada página web.

````python
import requests
from bs4 import BeautifulSoup

headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:100.0) Gecko/20100101 Firefox/100.0 "}

page = requests.get("http://dataquestio.github.io/web-scraping-pages/simple.html", headers = headers)
soup = BeautifulSoup(page.content, 'html.parser')
soup
````

#### Prettify

Podemos agora imprimir, devidamente formatado, o conteúdo HTML da página web, usando o método `prettify` em cima do objeto `BeautifulSoup`.

````python
print(soup.prettify())
````

#### get_text()

A função `get_text()` vai permitir obter todo e qualquer texto compreendido entre tags.

````python
soup.get_text()
````

````python
soup.text
````

Na eventualidade de pretendermos obter apenas texto de tags específicas teremos que recorrer aos: (1) métodos `find_all`; (2) aos `CSS selectors`.

##### Find_All

O método `find_all` como pôde observar anteriormente visa obter todas as instâncias de uma tag de uma só vez.

###### Finding all instances of a tag at once

````python
soup = BeautifulSoup(page.content, 'html.parser')
ListOfTagContents = soup.find_all('p')
ListOfTagContents
````

Observe que `find_all` devolve uma lista (neste caso com um único elemento, uma vez que existe uma única tag `p`), pelo que para acedermos ao texto teremos que usar um ciclo for (neste caso uma list comprehension):

````python
ListOfTexts = []
for content in ListOfTagContents:
    ListOfTexts.append(content.get_text())
ListOfTexts
````

````python
for item in ListOfTagContents:
    print(item.get_text())
````

````python
#Código similar ao anterior (mas com List Comprehensions)
ListOfTexts = [content.get_text() for content in ListOfTagContents]
ListOfTexts
````

ou recorrer aos índices:

````python
ListOfTagContents[0].get_text()
````

###### Searching for tags by class and id

Classes e ids são usadas por estilos CSS para determinar quais os estilos a aplicar a determinados elementos HTML. Por forma a entender como podemos obter valores de objetos mais complexos vamos proceder ao download e ao parse da seguinte página:

````python
headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:100.0) Gecko/20100101 Firefox/100.0 "}

page = requests.get("http://dataquestio.github.io/web-scraping-pages/ids_and_classes.html", headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
````

Agora, podemos usar o método `find_all` e posteriormente ao `get_text` para obter o texto de todas as tags `p` que tenham a classe `outer-text`:

````python
ListOfTagContents = soup.find_all('p', class_='outer-text')
contents = [content.get_text() for content in ListOfTagContents]
contents
````

O código seguinte permitiria posteriormente eliminar todos os caracteres desnecessários.

````python
for content in contents:
    contentList = content.split('\n')
    for text in contentList:
        if len(text) > 0:
            print(text.strip())
````

Já o exemplo seguinte permitiria obter o texto de qualquer tag que tivesse a class `outer-text` sem necessidade de ter a tag `p` associada:

````python
ListOfTagContents = soup.find_all(class_="outer-text")
[content.get_text() for content in ListOfTagContents]
````

Enquanto o exemplo seguinte permitiria obter o texto de todos os elementos que tivessem o `id=first`

````python
ListOfTagContents = soup.find_all(id="first")
[content.get_text() for content in ListOfTagContents]
````

###### Attributes

Os atributos conferem informação adicional a elementos HTML através de um par `name/value` (`name="value"`). Exemplos de atributos são o: 
- `href`:  `<a href="https://www.w3schools.com">This is a link</a>`
- `src`:  `<img src="img_girl.jpg">`
- `width` e `height`: `<img src="img_girl.jpg" width="500" height="600">`

O código seguinte exemplifica como é que o BeautifulSoup procede à extração de valores a partir de atributos. Neste exemplo vamos obter os links especificados no atributo `href`.

````python
import requests
from bs4 import BeautifulSoup

headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:100.0) Gecko/20100101 Firefox/100.0 "}

url = "https://arquivo.pt/noFrame/replay/20220127015829/http://www.ccc.ipt.pt/~ricardo/links.html"
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
ListOfContents = soup.find_all('a')
[a.get('href') for a in ListOfContents]
````

Ou em alternativa (note que `href=True` significa que vai obter todos os links se e só se eles estiverem presentes, caso contrário passa para o próximo).

````python
import requests
from bs4 import BeautifulSoup

headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:100.0) Gecko/20100101 Firefox/100.0 "}

url = "https://arquivo.pt/noFrame/replay/20220127015829/http://www.ccc.ipt.pt/~ricardo/links.html"
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
ListOfContents = soup.find_all('a',href=True)
[a['href'] for a in ListOfContents]
````

O código seguinte faz uso do `urljoin` para garantir que todos os URLs obtidos ficam com endereço absoluto (e.g. `http://www.ccc.ipt.pt/~ricardo/cv.html`) ao invés de relativo (e.g., `cv.html`)

````python
import requests
from urllib.parse import urljoin
from bs4 import BeautifulSoup

headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:100.0) Gecko/20100101 Firefox/100.0 "}

url = "https://arquivo.pt/noFrame/replay/20220127015829/http://www.ccc.ipt.pt/~ricardo/links.html"
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
ListOfContents = soup.find_all('a',href=True)
[urljoin(url,a['href']) for a in ListOfContents]
````

##### Using CSS Selectors

Ao invés de recorrermos ao `find_all` podemos procurar por itens usando `CSS` selectors (forma como a linguagem `CSS` permite aos desenvolvedores especificar tags `HTML` para estilos). Objetos `BeautifulSoup` suportam a pesquisa de uma página com recurso a seletores `CSS` por via do método `select`. Para melhor entendermos este processo vamos começar por obter o contéudo de uma página:

````python
import requests
from bs4 import BeautifulSoup

headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:100.0) Gecko/20100101 Firefox/100.0 "}

page = requests.get("http://dataquestio.github.io/web-scraping-pages/ids_and_classes.html", headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
soup.prettify
````

Os exemplos abaixo ilustram o código que permitiria obter texto que se encontre dentro de tags específicas:
- `p a` — devolve todas as tags `a` que se encontrem dentro de uma tag `p`.
- `body p a` — devolve todas as tags `a` que se encontrem dentro de uma tag `p` que por sua vez deverá estar dentro de uma tag `body`.
- `html body` — devolve todas as tags `body` que se encontrem dentro de uma tag `html`.
- `p.outer-text` — devolve todas as tags `p` com uma classe `outer-text`.
- `p#first` — devolve todas as tags `p` com um id `first`.
- `body p.outer-text` — devolve todas as tags `p` com uma classe `outer-text` que se encontrem dentro de uma tag `body`.

Para devolver o texto existente dentro da classe `outer-text` da tag `p` recorremos ao código seguinte:

````python
#select
ListOfTagContents = soup.select("p.outer-text")
ListOfTagContents
````

`ListOfTagContents` nada mais é do que uma lista, de duas posições, cada tal com uma tag `p.outer-text`. Para obtermos o texto de cada uma recorremos à seguinte list comprehension e ao método `get_text`.

````python
[content.get_text() for content in ListOfTagContents]
````

No exemplo abaixo recorremos a `select` para devolver o texto que se encontra dentro todas as tags `<div><p>`.

````python
ListOfTagContents = soup.select("div p")
ListOfTagContents
````

````python
[content.get_text() for content in ListOfTagContents]
````

````python
import requests
from bs4 import BeautifulSoup

headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:100.0) Gecko/20100101 Firefox/100.0 "}

page = requests.get("http://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168", headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')

ListOfTagPeriods = soup.select("div.tombstone-container p.period-name")
ListOfPeriods = [period.get_text() for period in ListOfTagPeriods]

ListOfTagTemp = soup.select("div.tombstone-container p.temp")
ListOfTemp = [temp.get_text() for temp in ListOfTagTemp]

ListOfTagShortDesc = soup.select("div.tombstone-container p.short-desc")
ListOfShortDesc = [ShortDesc.get_text() for ShortDesc in ListOfTagShortDesc]

ListOfTagDesc = soup.select("div.tombstone-container img")
ListOfDesc = [desc["title"] for desc in ListOfTagDesc]
````

De seguida vamos encapsular este código numa função. Em concreto esta função recebe um url de um produto da worten e retorna o respetivo preço.

````python
import requests
from bs4 import BeautifulSoup

def check_price(url):
    headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:100.0) Gecko/20100101 Firefox/100.0 "}
    
    page = requests.get(url, headers=headers)
    soup = BeautifulSoup(page.content, 'html.parser')
    price_integer = soup.find('span', class_="integer").get_text()
    price_decimal = soup.find('span', class_="decimal").get_text()
    price = float(f"{price_integer}.{price_decimal}")
    return price
````

Na posse do 16-digit code crie um ficheiro denominado `keys.json` com a seguinte estrutura:
```
{'gmail':['...@gmail.com', '16-digit passcode']}
```

````python
import json
import smtplib
from email.mime.text import MIMEText

def send_email(subject, body, sender, recipients, password):
    msg = MIMEText(body)
    msg['Subject'] = subject
    msg['From'] = sender
    msg['To'] = ', '.join(recipients)
    smtp_server = smtplib.SMTP_SSL('smtp.gmail.com', 465)
    smtp_server.login(sender, password)
    smtp_server.sendmail(sender, recipients, msg.as_string())
    smtp_server.quit()
    
def read_gmail_credentials():
    with open("keys.json", "r") as readfile:
        data = json.load(readfile)
    
    return data

url = "https://www.worten.pt/telemoveis-e-pacotes-tv/telemoveis-e-smartphones/smartphone-samsung-galaxy-a13-5g-6-5-4-gb-128-gb-preto-7640094"
price = check_price(url)
if price < 2000:
    #get gmail account and 16-pass code
    data = read_gmail_credentials()
    sender = data['gmail'][0]
    password = data['gmail'][1]

    subject = "Worten: Price Notification"
    body = f"Price changed {str(price)}.\n Check the URL: {url}"

    #Specify the recipient email(s)
    recipients = ["ricardo.campos@ubi.pt"]

    send_email(subject, body, sender, recipients, password)
    print("email sent")
else:
    print("email was not sent")

````

````python
import time
from random import randint
import requests
from bs4 import BeautifulSoup


#Determinar o num de pages
url = "https://www.imovirtual.com/comprar/apartamento/alvor-portimao/?search%5Bregion_id%5D=8&search%5Bsubregion_id%5D=112&search%5Bcity_id%5D=8399873&search%5Border%5D=quality_score&page=1"
page = requests.get(url, headers={"User-Agent": "Defined"})
soup = BeautifulSoup(page.content, 'html.parser')
n_pages = int(soup.find('div', class_="offers-index pull-left text-nowrap").get_text().split(':')[-1].strip())
if n_pages % 24 == 0:
    n_pages = int(n_pages/24)
else:
    n_pages = int(n_pages/24) + 1

print(n_pages)
````

````python
import time
from random import randint
import requests
from bs4 import BeautifulSoup

houses = []
numHouses = 0

for page in range(1, n_pages + 1):
    
    url = f"https://www.imovirtual.com/comprar/apartamento/alvor-portimao/?search%5Bregion_id%5D=8&search%5Bsubregion_id%5D=112&search%5Bcity_id%5D=8399873&search%5Border%5D=quality_score&page={page}"
    print(url)
    
    page = requests.get(url, headers={"User-Agent": "Defined"})
    soup = BeautifulSoup(page.content, 'html.parser')

    house_containers = soup.find_all('article')
    if house_containers != []:
        for house in house_containers:
            numHouses += 1

            # title
            title = house.find('span', class_="offer-item-title").get_text().strip()


            # price and status
            try:
                price = float(house.find('li', class_="offer-item-price").get_text().replace(u'\xa0', u' ').replace('€', '').replace('.', '').replace(' ', ''))
                price = round(price, 2)
            except:
                price = None



            #Area
            try:
                area = int(house.find('li', class_="hidden-xs offer-item-area").get_text()[:-3])
                pricem2 = round(price/area, 2)
            except:
                area = None
                pricem2 = None

            # URL and Photo link

            url = house.find('a').get('href')
            try:
                img = house.find('span', class_="img-cover lazy").get('data-src')
            except:
                img = None

            houses.append({'RealEstateAgency':'Imovirtual', 'title': title, 'price': price, 'area': area, 'url': url, 'img': img})
    else:
        break

    time.sleep(randint(1, 5))
````

#### Locating Elements

Uma vez aberto o browser, necessitamos especificar os próximos passos. Com vista a esse objetivo necessitamos detetar o target da página web e para isso necessitamos mais uma vez de recorrer ao `Inspect Elements`. Para identificar certos elementos da página temos várias hipóteses:

https://selenium-python.readthedocs.io/locating-elements.html

- ByID
- ByName
- ByClassName
- ByTagName

Comece por considerar o seguinte código HTML:

```
<html> 
 <body> 
  <form id="loginForm"> 
   <input name="username" type="text" /> 
   <input name="password" type="password" /> 
   <input name="continue" type="submit" value="Login" /> 
  </form> 
 </body> 
<html> 
```

##### ByID

Se tivermos o ID exato do elemento que estamos à procura podemos recorrer ao `By.ID`. Por exemplo, para termos acesso ao formulário de login poderiamos recorrer ao seguinte código: `login_form = browser.find_element(By.ID,'loginForm')`

##### ByName

Se tivermos um objeto `name` podemos recorrer ao `By.NAME`.  Por exemplo, para termos acesso à caixa do username: `element = browser.find_element(By.NAME,'username')`.

##### ByClassName

Se tivermos o `name` da `class` podemos recorrer ao `By.CLASS_NAME`. Considere o seguinte código HTML:

```
<html> 
 <body> 
  <p class="content">Site content goes here.</p> 
</body> 
<html> 
```

Por exemplo: `element = browser.find_element(By.CLASS_NAME, 'content')`.



##### ByTagName

Se tivermos o `tag` podemos recorrer ao `By.TAG_NAME`. Considere o seguinte código HTML:

```
<html> 
 <body> 
  <h1>Welcome</h1> 
  <p>Site content goes here.</p> 
</body> 
<html> 
```

Por exemplo: `element = browser.find_element(By.TAG_NAME('h1')`.

##### ByCSSSelector

Use this when you want to locate an element using CSS selector syntax. With this strategy, the first element matching the given CSS selector will be returned. If no element matches the provided CSS selector, a NoSuchElementException will be raised.

```
<html>
 <body>
  <p class="content">Site content goes here.</p>
</body>
</html>
```

Por exemplo: `element = browser.find_element(By.CSS_SELECTOR, 'p.content')`

#### Getting Elements by XPath

Nem todos os elementos têm um ID ou talvez não queira aceder às tags HTML. Existem no entanto outras formas de devolver elementos específicos, através do `XPath`, que nada mais é do que um XML path, uma sintaxe ou linguagem para encontrar qualquer elemento numa página web através de uma expressão XML path usando a estrutura HTML DOM. No fundo a XPath percorre a hierarquia do elemento root da página web até ao elemento desejado.

Para melhor entender este processo considere o seguinte código HTML:

```
<html> 
 <body> 
  <form id="loginForm"> 
   <input name="username" type="text" /> 
   <input name="password" type="password" /> 
   <input name="continue" type="submit" value="Login" /> 
  </form> 
 </body> 
<html> 
```

Para selecionar o objeto `form` basta especificar o seguinte caminho: `/html/body/form[1]`. Dado que poderão existir vários forms na página web, recorre-se ao índice 1 para indicar que pretendemos aceder apenas ao primeiro. O código completo encontra-se abaixo:

```
login_form = browser.find_element_by_xpath("/html/body/form[1]")
```

Naturalmente, se tivessemos escrito apenas `/html/body/form` estaríamos na presença de uma lista (neste caso com apenas um único elemento) com elementos selenium sobre os quais poderiamos iterar. Nesse caso teríamos ter que ter usado o `find_elements_by_path`.

##### Syntax

A estrutura básica de um XPath no entanto pode ser um pouco mais complexa do que indicar apenas tagnames. A figura seguinte, exemplifica a sintaxe de um XPath:

<img src="https://www.di.ubi.pt/~rcampos/assets/img_tutorials/Text2Analytics/XPath.jpg" width="500" height="500" align="center"/>

<center>https://www.guru99.com/xpath-selenium.html</center>

- //: Seleciona o nó atual;
- Tagname: Tagname do nó em particular;
- @: Seleciona o atributo;
- Attribute: Nome do atributo do nó;
- Value: Valor do atributo.

Os atributos poderão ser `IDs`, `ClassName`, `Name` entre muitas outras possibilidades.

##### Types of XPath

Existem dois tipos de path: (1) Absolutos; (2) Relativos.

Um path <b>Absoluto</b> é a forma direta de encontrar um elemento. Naturalmente se houver uma alteração feita na estrutura da página, a obtenção do elemento XPath poderá falhar, o que pode ser visto como uma desvantagem. O elemento central deste tipo de path é que o mesmo inícia com (`/`), o que significa que a seleção tem início no nó de root.

Um path <b>Relativo</b> por sua vez inicia com (`//`), o que significa que pode começar a procura a partir de qualquer parte da página web, evitando a necessidade de especificar a totalidade do path. A obtenção de um path relativo pode significar no entanto um processo bastante complexo. Para simplificar, vamos proceder à instalação de um addon que permite obter o path relativo de forma automática. No caso do Firefox recorremos ao [ChroPath](https://addons.mozilla.org/pt-PT/firefox/addon/chropath-for-firefox/). Para o Chrome pode recorrer a este [addon](https://chrome.google.com/webstore/detail/chropath/ljngjbnaijcbncmcnjfhigebomdlkcjo). Uma vez instalado, podemos agora obter path absolutos e relativos sem qualquer dificuldade, bastando para tal inspecionar a página web e selecionar o objeto HTML correspondente.

Para melhor exemplificarmos este processo recorremos à página web `http://demo.guru99.com/test/selenium-xpath.html` e selecionamos a área do UserID.

- //input[@type='text']				
- //label[@id='message23']
- //input[@value='RESET']
- //*[@class='barone']
- //a[@href='http://demo.guru99.com/']

######  Radio Popular (Infinite Scroll Down)

O próximo código permite fazer scroll down infinito ao longo da página web de telemóveis da [Rádio Popular](https://www.radiopopular.pt/categoria/telemoveis-e-smartphones/#), uma operação muito comum em redes sociais como é o caso do Facebook, Linkedin, Youtube, etc.

Com vista a esse objetivo definimos a função `scroll_down` que recorre a código javascript para determinar o fim da página web. Se é verdade que nas redes sociais o scroll-down é habitualmente never-ending, tal não acontece na página da Rádio Popular, onde existe um fim de página correspondente ao último telemóvel do catálogo a ser mostrado aos utilizadores. 

Em concreto, o código da função `scroll_down` começa por determinar a altura, em pixeis, de um dado elemento, no caso em concreto do documento que está a ver no browser (`return document.body.scrollHeight`). **Nota**: * scrollHeight value is equal to the minimum height the element would require in order to fit all the content (e por "all the content" entenda-se todo o conteúdo presentemente carregado no browser) in the viewport without using a vertical scrollbar. Apesar de todos os browsers poderem seguir o mesmo procedimento, diferentes browsers podem apresentar valores ligeiramente distintos. Naturalmente este valor é também ele dependente da altura (resolução) do seu próprio ecrã.

O código entra de seguida num ciclo `while true` que apenas será interrompido (`break`) no momento em que, por não haver mais elementos a carregar, o `new_height` é igual ao `last_height`.

Terminado o scroll down o sistema guarda o resultado da colecta num ficheiro `data/rp.html` e fecha o browser.

````python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import time
from random import randint
import os

path = "data"

def scroll_down(driver):
    """A method for scrolling the page."""

    # Get scroll height.
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:

        # Scroll down to the bottom.
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load the page.
        time.sleep(randint(1,3))

        # Calculate new scroll height and compare with last scroll height.
        new_height = driver.execute_script("return document.body.scrollHeight")

        if new_height == last_height:
            break
        last_height = new_height



options = webdriver.ChromeOptions()
options.add_argument("--disable-infobars")
options.add_argument("start-maximized")
options.add_argument("--disable-extensions")

# Pass the argument 1 to allow and 2 to block
options.add_experimental_option("prefs", {
    "profile.default_content_setting_values.notifications": 1
})

# chromedriver should be in the same folder as file
browser = webdriver.Chrome(options=options)
browser.implicitly_wait(5)

# Open login page
browser.get('https://www.radiopopular.pt/categoria/telemoveis-e-smartphones/#')
time.sleep(randint(1,3))
browser.maximize_window()

# Get scroll height
scroll_down(browser)

company_page = browser.page_source

soup = bs(company_page, "html.parser")

if not os.path.exists(path):
    os.makedirs(path)
        
with open(f'{path}/rp.html', "w", encoding="utf-8") as file:
    file.write(str(soup.prettify()))

browser.close()
````

Uma alternativa ao código anterior (em termos de `scroll_down`) passa pelo recurso à utilização da tecla `PAGE_DOWN`. Naturalmente o recurso a esta tecla é infinito, a menos que se estipule uma condição de interrupção. No caso da Rádio Popular a condição decorre do facto de o número de telemóveis carregados na página ser igual ao número total de elementos a listar. Nota: ambas as informações estão disponíveis no todo da página web. 

````python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as bs
import time
from random import randint
import os

path = "data"

def scroll_down(driver):
    """A method for scrolling the page."""
    total = driver.find_element(By.XPATH, '//*[@id="total"]')
    total = int(total.text)

    #page = driver.find_element_by_tag_name('body')
    page = driver.find_element(By.TAG_NAME, 'body')

    while True:
        time.sleep(randint(1,3))
        numLoaded = driver.find_element(By.XPATH, '//*[@id="products"]')
        numLoaded = int(numLoaded.text)
        if numLoaded == total:
            break
        else:
            page.send_keys(Keys.PAGE_DOWN)


# accessing Chromedriver
options = webdriver.ChromeOptions()
options.add_argument("--disable-infobars")
options.add_argument("start-maximized")
options.add_argument("--disable-extensions")

# Pass the argument 1 to allow and 2 to block
options.add_experimental_option("prefs", {
    "profile.default_content_setting_values.notifications": 1
})

# chromedriver should be in the same folder as file
browser = webdriver.Chrome()
browser.implicitly_wait(5)

# Open login page
browser.get('https://www.radiopopular.pt/categoria/telemoveis-e-smartphones/#')
time.sleep(randint(1,3))
browser.maximize_window()

# Get scroll height
scroll_down(browser)

company_page = browser.page_source

soup = bs(company_page, "html.parser")

if not os.path.exists(path):
    os.makedirs(path)

with open(f'{path}/rp.html', "w", encoding="utf-8") as file:
    file.write(str(soup.prettify()))

browser.close()
````

###### Scientist Search Results - Howard Hughes Medical Institute

O código seguinte exemplifica como obter informação a partir da página [Scientist Search Results](https://www.hhmi.org/scientists/browse) a qual obfusca o código javascript devolvido. Em primeiro lugar começamos por obter o número total de resultados listados na página web (n_res). Tendo em conta que cada página web mostra informação sobre 8 investigadores obtemos depois o número total de páginas (n_pages) sobre as quais deve ser feito o processo de scraping.

Observe que recorremos a `XPATH` para clicar no botão de cookies. De seguida identificamos o `shadow-root` principal (`hhmi-search-module`) e o elemento que pretendemos (`hhmi-search-nav`), a partir do qual poderá obter com recurso ao método `get_attribute` o `results`.

````python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.by import By

browser = webdriver.Chrome()

browser.get("https://www.hhmi.org/search/people?sort_by=field_name_family&sort_order=ASC&page=0")
browser.find_element(By.XPATH, "//button[contains(text(),'CLOSE')]").click() #para clicar bo botão close...cookies esses que apenas aparecem quando executado a partir do selinium    

shadow_host1 = browser.find_element(By.CSS_SELECTOR, "hhmi-search-module")
shadow_root1 = browser.execute_script('return arguments[0].shadowRoot', shadow_host1)

shadow_host2 = shadow_root1.find_element(By.CSS_SELECTOR, "hhmi-search-nav")
r_res = shadow_host2.get_attribute('results').split()[0]

#Get the number of pages to scrap
if n_res % 8 == 0:
    n_pages = int(n_res/8)
else:
    n_pages = int(n_res/8) + 1

print(n_pages)
browser.close()
````

De seguida vamos fazer um ciclo `for` para iterar ao longo das várias páginas que incluem a lista de investigadores. Para isso vamos mais uma vez, e para cada página, identificar o `shadow-root` principal (`hhmi-search-module`) e o elemento que pretendemos (`hhmi-person-search-result-group`), a partir do qual poderá obter com recurso ao método `get_attribute` os vários itens relacionados com os investigadores. A cada iteração, adicionámos à `ListOfResearchers` um dicionário com os vários elementos coletados.

Note que a identificação dos `shadow-root` é feita através do inspecionar elementos olhando para a linha que aparece acima, precisamente, da identificação `shadow-root`.

````python
browser = webdriver.Chrome()

ListOfResearchers = []

for i in range(0,n_pages):
    browser.get(f"https://www.hhmi.org/search/people?sort_by=field_name_family&sort_order=ASC&page={i}")
    
    shadow_host1 = browser.find_element(By.CSS_SELECTOR, "hhmi-search-module")
    shadow_root1 = browser.execute_script('return arguments[0].shadowRoot', shadow_host1)
    
    shadow_host2 = shadow_root1.find_element(By.CSS_SELECTOR, "hhmi-person-search-result-group")   
    ListOfTagContents = shadow_host2.find_elements(By.CSS_SELECTOR, 'hhmi-card')
    for element in ListOfTagContents:
        try:
            name = element.get_attribute('heading').split(",")[0].strip()
        except:
            name = None
        
        try:
            category = element.get_attribute('heading').split(",")[1].strip()
        except:
            category = None
        
        try:
            role = element.get_attribute('subtitle').strip()
        except:
            role = None
        
        try:
            time = element.get_attribute('label').split(" · ")[0]
        except:
            time = None
        
        try:
            url = "https://www.hhmi.org" + element.get_attribute('linkurl')
        except:
            url = None
            
        try:
            img = "https://www.hhmi.org" + element.get_attribute('imgsrc')
        except:
            img = None
        
        try:
            affiliation = element.get_attribute('label').split(" · ")[1].strip()
        except:
            affiliation = None
            
        ListOfResearchers.append({"name":name,"category":category,"role":role, "time":time, "affiliation":affiliation, "url":url,"img":img})
        
    print(f"https://www.hhmi.org/search/people?sort_by=field_name_family&sort_order=ASC&page={i}")

browser.close()
````

Com vista a colocar os dados num dataframe começamos por definir a seguinte função a qual permite customizar a saída do dataframe. Em particular, vamos formatar o url para que ele apareça clicável e o nome para que ele apareça associado ao url da página web do investigador.

````python
def CustomizeDataframe(df):
    def make_visible_img(url):
        return f'<img src="{url}" style="max-height:124px;"></img>'

    def make_clickable_link(url, name):
        return f'<a target="_blank" href="{url}">{name}</a>'

    df['image'] = df.apply(lambda x: make_visible_img(x['img']), axis=1)
    df['name'] = df.apply(lambda x: make_clickable_link(x['url'], x['name']), axis=1)

    return df
````
